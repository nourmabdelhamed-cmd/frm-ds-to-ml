## Production Does Not Mean Heavy CI

CI should answer a narrow question:

**Did this change break the workflow contract?**

It does not need to download VOC or train a real segmentation model on every push.

::: {.notes}
Timing: 4 minutes. Keep the CI story pragmatic: fast tests make people use CI; slow fragile checks get ignored.
:::

## What The Tests Cover

- `params.yaml` is parsed into a real config object
- synthetic batches have image and mask shapes expected by segmentation training
- the tiny model returns `[batch, classes, height, width]`
- one training step writes a checkpoint
- `train.py` accepts config and command-line overrides

The shift is from **my experiment works** to **the team can trust and reproduce it**.

::: {.notes}
Keep reminding the audience that the metric is not accuracy. The metric is whether another person can understand, rerun, and safely change the workflow.
:::
