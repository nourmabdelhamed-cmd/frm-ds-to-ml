## The Next Problem

Now several people can run training. That creates a new question:

**Which run produced which checkpoint, with which config, and how do we compare them?**

This is where MLflow enters the workflow.

::: {.notes}
Timing: 4 minutes. Emphasize run history and shared memory first; model registry second as the lifecycle extension.
:::

## What We Track

```python
mlflow.set_experiment(config.mlflow_experiment)

with mlflow.start_run(run_name=config.run_name):
    mlflow.log_params(config)
    mlflow.log_metric("train_loss", train_loss, step=epoch)
    mlflow.log_artifact(checkpoint_path)
```

Tracking makes training inspectable after the terminal output is gone.

## Optional Model Registry Hook

The config can opt into model logging and registration:

```yaml
tracking:
  mlflow_experiment: ipcv-voc-demo
  checkpoint_path: models/voc-tiny-unet.pt
  log_model: true
  registered_model_name: ipcv-voc-segmentation
```

For the beginner demo, this is off by default. The workflow supports it when the tracking backend is ready.

## Development Loop With MLflow

```{mermaid}
flowchart LR
  A[Change config] --> B[Run train.py]
  B --> C[Log params and metrics]
  C --> D[Compare runs]
  D --> E[Choose checkpoint]
  E --> F[Optionally register model]
```

## Why This Matters In A Team

- You can compare runs without reading every notebook
- You can explain where a checkpoint came from
- You can keep failed runs visible
- You can promote a model through a lifecycle later
